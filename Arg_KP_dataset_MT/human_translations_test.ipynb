```python
def replace_topics(df):
    # Get unique topics from the dataframe
    unique_topics = df['topic'].unique()

    # Define the new topics in Greek
    new_topics = [
        'Οι παιδικοί εμβολιασμοί ρουτίνας θα πρέπει να είναι υποχρεωτικοί',
        'Οι πλατφόρμες κοινωνικής δικτύωσης θα πρέπει να ρυθμίζονται από την κυβέρνηση',
        'Οι ΗΠΑ είναι μια καλή χώρα να ζει κανείς'
    ]

    # Create a mapping dictionary for old topics to new topics
    topics_mapping = dict(zip(unique_topics, new_topics))

    # Replace the old topics with the new topics in the dataframe
    df['topic'] = df['topic'].map(topics_mapping)

    return df
```

```python
def translate_kp_dataset(file_path):
    # Read the original CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Apply the translation functions
    df = replace_kps(df)
    df = replace_topics(df)

    # Save the translated DataFrame to a new CSV file
    df.to_csv('key_points_translated_test.csv', index=False, encoding='utf-8')

    return df
```

```python
def replace_args(df, txt_file):
    # Read the arguments from the text file
    with open(txt_file, 'r', encoding='utf-8') as file:
        new_arguments = file.readlines()

    # Strip newline characters from each argument
    new_arguments = [arg.strip() for arg in new_arguments]

    # Ensure the number of arguments matches the number of rows in the dataframe
    if len(new_arguments) != len(df):
        raise ValueError("The number of arguments in the text file does not match the number of rows in the dataframe")

    # Replace the 'argument' column with the new arguments
    df['argument'] = new_arguments

    return df

def translate_argument_dataset(file_path, txt_file, output_file):
    # Read the original CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Apply the argument replacement function
    df = replace_args(df, txt_file)
    df = replace_topics(df)
    # Save the translated DataFrame to a new CSV file
    df.to_csv(output_file, index=False)

    return df
```

# Correct capitalisation

```python
import re
original_df = pd.read_csv('/content/drive/MyDrive/Πτυχιακή/Datasets/madlad400-3b/arguments_test.csv')
translated_df = pd.read_csv('/content/drive/MyDrive/Πτυχιακή/Datasets/madlad400-3b/arguments_translated_test.csv')
def adjust_capitalization(original_text, translated_text):
    if not original_text or not translated_text:
        return translated_text  # Handle cases where the text might be empty or None

    # Check if the first character in the original text is uppercase
    original_first_is_upper = original_text[0].isupper()

    # Apply the same capitalization to the translated text
    if original_first_is_upper:
        translated_text = translated_text[0].upper() + translated_text[1:]
    else:
        translated_text = translated_text[0].lower() + translated_text[1:]

    return translated_text
```

```python
# Ensure both DataFrames have the same number of rows
assert len(original_df) == len(translated_df), "The datasets do not have the same number of rows"

# Apply the capitalization adjustment function
translated_df['argument'] = translated_df.apply(
    lambda row: adjust_capitalization(original_df.loc[row.name, 'argument'], row['argument']),
    axis=1
)

# Save the adjusted translated DataFrame to a new CSV file
translated_df.to_csv('translated_adjusted.csv', index=False)
```

```python
def adjust_text(original_text, translated_text):
    if not original_text or not translated_text:
        return translated_text  # Handle cases where the text might be empty or None

    # Adjust capitalization
    original_first_is_upper = original_text[0].isupper()
    if original_first_is_upper:
        translated_text = translated_text[0].upper() + translated_text[1:]
    else:
        translated_text = translated_text[0].lower() + translated_text[1:]

    # Adjust punctuation
    original_ends_with_dot = original_text.endswith('.') and not original_text.endswith('...')
    translated_ends_with_dot = translated_text.endswith('.') and not translated_text.endswith('...')

    if original_ends_with_dot and not translated_ends_with_dot:
        translated_text += '.'
    elif not original_ends_with_dot and translated_ends_with_dot:
        translated_text = translated_text.rstrip('.')

    return translated_text
```
